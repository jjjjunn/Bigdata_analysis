```python
# 데이터 로드
import pandas as pd

df = pd.read_csv(DataUrl)
# print(df.head(3))

# quantity 컬럼 값이 3인 데이터를 추출하여 첫 5행 출력
qty_3 = df['quantity'] == 3
# print(df[qty_3].head())

# quantity 컬럼 값이 3인 데이터를 추출하여 index를 0부터 정렬하고 첫 5행 출력
df_qty3 = df[qty_3].reset_index(drop=True) # drop_True 넣지 않으면 기존 인덱스 출력됨
# print(df_qty3.head())
# print(help(df.sort_values))

# quantity, item_price 두 개의 컬럼으로 구성된 새로운 데이터 프레임 정의
new_df = df[['quantity', 'item_price']]
# print(new_df.head())

# item_price 컬럼의 달러표시 문자 제거하고 float 타입으로 저장하여 new_price 컬럼에 저장
df['new_price'] = df['item_price'].str.replace('$', '').astype('float64')
# print(df['new_price'])

# new_price 컬럼이 5이하의 값을 가지는 데이터프레임을 추출하고, 전체 갯수 구하기
cond = df['new_price'] <= 5
# print(df[cond])
# print(len(df[cond]))

# item_name 명이 Chiken Salad Bowl 인 데이터 프레임을 추출하고 index 값 초기화
cond2 = df['item_name'] == 'Chicken Salad Bowl'
# print(df[cond2])
new_df2 = df[cond2].reset_index(drop=True)
# print(new_df2.head())

# new_price 값이 9이하이고 item_name 값이 Chicken Salad Bowl 인 데이터 프레임 추출
cond3 = df['new_price'] <= 9

new_df3 = df[cond2 & cond3]
# print(new_df3.head())

# df의 new_price 컬럼 값에 따라 오름차순으로 정리하고 index 초기화
df = df.sort_values('new_price').reset_index(drop=True)
# print(df.head())

# df의 item_name 컬럼 값 중 Chips 포함하는 경우의 데이터 출력
cond4 = df['item_name'].str.contains('Chips')
# print(df[cond4].head(10))

# df의 짝수번째 컬럼만을 포함하는 데이터프레임 출력
# print(df.iloc[:,::2]) # : 모든 행, ::n 스텝 -> 열을 n 칸 간격으로 선택

# df의 new_price 컬럼 값에 따라 내림차순으로 정리하고 index 초기화
df = df.sort_values('new_price', ascending=False).reset_index(drop=True)
# print(df.head(10))

# df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터 인덱싱
cond1 = df['item_name'] == 'Steak Salad'
cond2 = df['item_name'] == 'Bowl'
# print(df[cond1 | cond2])

# df의 item_name 컬럼 값이 Steak Sald 또는 Bowl 인 데이터를 데이터 프레임화 한 후, item_name을 기준으로 중복행이 있으면 제거하되 첫번째 케이스만 남기기
new_df = df[cond1 | cond2]
# print(help(df.drop))

# print(new_df.drop_duplicates('item_name')) # 중복 제거: drop_duplicates('기준 컬럼명')

# print(help(df.drop_duplicates))
# df의 item_name 컬럼 값이 Steak Sald 또는 Bowl 인 데이터를 데이터 프레임화 한 후, item_name을 기준으로 중복행이 있으면 제거하되 마지막 케이스만 남기기
# print(new_df.drop_duplicates('item_name', keep='last')) # keep='last': 마지막 케이스만 남기기

# df 데이터 중 new_price 값이 new_price 값의 평균값 이상을 가지는 데이터 인덱싱
mean = df['new_price'].mean()
# print(mean) # 7.464335785374297

cond = df['new_price'] >= mean
# print(df[cond])

# df의 데이터 중 item_name 값이 lzze 데이터를 Fizzy Lizzy로 수정
cond = df['item_name'] == 'Izze'
# print(df[cond])
df['item_name'] = df['item_name'].str.replace('Izze', 'Fizzy Lizzy')
# print(df['item_name'] == 'Fizzy Lizzy')

# print(df.head(10))

# df 데이터 중 choice_description 값이 NaN인 데이터 개수 구하기
# print(df['choice_description'].isnull().sum())

# df 데이터 중 choice_description 값이 NaN인 데이터를 NoData 값으로 대체 (loc 이용)
# print(df['choice_description'].fillna('NoData'))
# print(help(df.loc))

# df.loc['조건', '컬럼'] = '할당값'
df.loc[df.choice_description.isnull(), 'choice_description'] = 'NoData'
# print(df.head(10))

# df 데이터 중 choice_description 값에 Black 이 들어가는 경우를 인덱싱
cond = df['choice_description'].str.contains('Black')
# print(df[cond])

# df 데이터 중 choice_description 값에 Vegetables 가 들어가지 않는 경우의 개수 출력
cond = df['choice_description'].str.contains('Vegetables')
# print(df[~cond])
# print(len(df[~cond]))

# df의 데이터 중 item_name 값이 'N'으로 시작하는 데이터를 모두 추출
cond = df['item_name'].str.startswith('N')
# print(df[cond])
# print(len(df[cond])) # 개수: 27

# df 데이터 중 item_name 값의 단어갯수가 15개 이상인 데이터 인덱싱
# cond = len(df['item_name'].str()) >= 15 # 오답
# cond = len(str(df['item_name'])) >= 15 # 오답
cond = df.item_name.str.len() >= 15 # 특정 컬럼의 글자수 세기
# print(df[cond])

# df 데이터 중 new_price 값이 lst에 해당하는 경우의 데이터 프레임을 구하고 그 갯수 출력
lst = [1.69, 2.39, 3.39, 4.45, 9.25, 10.98, 11.75, 16.98]
import numpy as np
# print(help(np.isin))

cond = df['new_price'].isin(lst)
# print(df[cond])
print(len(df[cond]))
```
