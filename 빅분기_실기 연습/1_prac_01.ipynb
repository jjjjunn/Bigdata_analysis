# 문제16. 이상치, IQR


1.   'views' 컬럼의 1사분위수, 3사분위수, IQR 계산
2.   이상치 조건에 맞는 데이터 찾기
> (1q - (IQR * .15) 보다 작은 값 & 3q + (IQR * 1.5) 보다 큰 값
3.   이상치 데이터의 'views' 컬럼 합 정수로 출력


```python
import pandas as pd

df = pd.read_csv(type1_data1)
# print(df)

# 1.'views' 컬럼의 1사분위수, 3사분위수, IQR 계산
q1 = df['views'].quantile(.25)
q3 = df['views'].quantile(.75)
views_iqr = q3 - q1
# print(views_iqr) # 5442.5

# 2. 이상치 조건에 맞는 데이터 찾기 (1q - (IQR * 1.5) 보다 작은 값 & 3q + (IQR * 1.5) 보다 큰 값)
cut1 = q1 - (views_iqr * 1.5) # 5132.75
cut2 = q3 + (views_iqr * 1.5) # 16637.25

# print(cut1, cut2)

cond1 = df['views'] < cut1
cond2 = df['views'] > cut2

df = df[cond1 | cond2]
# print(df)

# 3. 이상치 데이터의 'views' 컬럼 합 정수로 출력
iqr_sum = int(df['views'].sum())
print(iqr_sum)
```

# 문제27. 시간 범위, 속도(km/h)
1. 점심시간(10시부터 13시 전)에 주문된 배달 데이터 출력
2. 점심시간 주문 건 중 과속(평균 속도가 50kh/h 이상)하는 주문 수를 정수로 출력
> 배달시간 = 실제도착시간 - 주문시간
> 속도(km/h) = 거리(km)/시간(h)

```python
import pandas as pd

df = pd.read_csv(del_time)
# print(df)

# 1. 점심시간(10시부터 13시 전)에 주문된 배달 데이터 출력
df['주문시간'] = pd.to_datetime(df['주문시간'])
df['order_h'] = df['주문시간'].dt.hour
# print(df)

df['점심여부'] = (df['order_h'] > 10) & (df['order_h'] < 13)
cond1 = df['점심여부'] == True
# print(df[cond1]) # 72 rows

#2. 점심시간 주문 건 중 과속(평균 속도가 50kh/h 이상)하는 주문 수를 정수로 출력
# > 배달시간 = 실제도착시간 - 주문시간
# > 속도(km/h) = 거리(km)/시간(h)

df['실제도착시간'] = pd.to_datetime(df['실제도착시간'])
df['배달시간'] = (df['실제도착시간'] - df['주문시간']).dt.total_seconds() / 3600
df['속도'] = df['거리'] / df['배달시간']
# print(df)

cond2 = df['속도'] >= 50

result = df[cond1 & cond2]
print(int(len(result)))
```

# 문제29. 함수, 월별 집계
1. 배달료 계산 기준표에 따라 각 주문에 대한 배달료 계산
2. 월별로 배달료의 총합 집계
3. 배달료가 가장 많이 발생한 월을 찾고, 그 월의 총 배달료를 정수로 계산

[배달료 기준표]
- 5km 미만: 2,000원
- 5km 이상 ~ 10km 미만: 4,000원
- 10km 이상 ~ 15km 미만: 6,000원
- 15km 이상 ~ 20km 미만: 8,000원

```python
import pandas as pd

df = pd.read_csv(del_time)
# print(df)

# 1. 배달료 계산 기준표에 따라 각 주문에 대한 배달료 계산
def del_fee(distance):
  if distance < 5:
    return 2000
  elif distance < 10:
    return 4000
  elif distance < 15:
    return 6000
  elif distance < 20:
    return 8000

df['배달료'] = df['거리'].apply(del_fee)
# print(df)

#2. 월별로 배달료의 총합 집계
df['주문시간'] = pd.to_datetime(df['주문시간'])
df['주문월'] = df['주문시간'].dt.month
# print(df)

monthly_fee = df.groupby('주문월')['배달료'].sum()
# print(monthly_fee)

# 3. 배달료가 가장 많이 발생한 월을 찾고, 그 월의 총 배달료를 정수로 계산
print(int(monthly_fee.max()))
```

# 문제37. 재구조화(unstack), 맵핑
1. 각 행에서 판매수량과 단가를 이용하여 매출액 계산
2. 요일을 평일과 주말로 구분하고, 매장별 평일과 주말 매출액 합계 계산
3. 매장별 평일과 주말 매출액 차이를 절대값으로 계산하고, 모든 매장 중 가장 큰 절대값 차이를 출력

```python
import pandas as pd

df = pd.read_csv(store)
# print(df)

# 1. 각 행에서 판매수량과 단가를 이용하여 매출액 계산
df['매출액'] = df['판매수량'] * df['단가']
# print(df)

# 2. 요일을 평일과 주말로 구분하고, 매장별 평일과 주말 매출액 합계 계산
day_mapping = {'월': '평일', '화': '평일', '수': '평일', '목': '평일', '금': '평일', '토': '주말', '일': '주말'}

df['주말여부'] = df['요일'].map(day_mapping)
# print(df)

store_sales_by_dw = df.groupby(['매장코드', '주말여부'])['매출액'].sum().unstack()
# print(store_sales_by_dw)

# 3. 매장별 평일과 주말 매출액 차이를 절대값으로 계산하고, 모든 매장 중 가장 큰 절대값 차이를 출력
store_sales_by_dw['차이'] = abs(store_sales_by_dw['주말'] - store_sales_by_dw['평일'])
# print(store_sales_by_dw)

print(store_sales_by_dw['차이'].max())
```

# 문제38. 피벗테이블
1. 각 Region과 Channel 조합별로 제품(Product) A, B의 총 판매액 계산
2. 제품 A의 매출 비율(A비율) 계산
> A비율 = (제품 A 판매액) / (제품 A 판매액 + 제품 B 판매액)
3. A 비율 중 최댓값을 찾아 반올림하여 소수 둘째자리까지 출력

```python
import pandas as pd

df = pd.read_csv(region_csv)
# print(df)

# 1. 각 Region과 Channel 조합별로 제품(Product) A, B의 총 판매액 계산
pivot = pd.pivot_table(df, index=['Region', 'Channel'], columns='Product', values='Sales', aggfunc='sum')
# print(pivot)

# 2. 제품 A의 매출 비율(A비율) 계산
#> A비율 = (제품 A 판매액) / (제품 A 판매액 + 제품 B 판매액)
pivot['총매출'] = pivot['A'] + pivot['B']
pivot['A비율'] = pivot['A'] / pivot['총매출']
# print(pivot)

# 3. A 비율 중 최댓값을 찾아 반올림하여 소수 둘째자리까지 출력
result = pivot['A비율'].max()
print(round(result, 2))
```

2. 위의 결과 중, 매출 합계(Sales)가 1400을 초과하는 경우가 몇 건인지 계산
