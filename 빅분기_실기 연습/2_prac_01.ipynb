# 이진분류

## 이진분류 연습1: 환자의 당뇨병 여부 예측
1. train과 test 데이터셋을 이용
2. 예측할 컬럼: outcome (0: 정상, 1:당뇨병)
3. 제출 파일: pred 예측값(당뇨병일 확률)을 포함한 d_result.csv
4. 제출한 모델의 성능은 ROC-AUC 평가지표에 따라 체점

```python
pt2_ch6 = "https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert_v2/refs/heads/main/part2/ch6"
dia_train = pt2_ch6 + "/diabetes_train.csv"
dia_test = pt2_ch6 + "/diabetes_test.csv"
```

```python
import pandas as pd

# 파일 불러오기
d_train = pd.read_csv(dia_train)
d_test = pd.read_csv(dia_test)

# 데이터 크기
# print(d_train.shape, d_test.shape) # (614, 9) (154, 8)

# 데이터 일부 보기
# print(d_train.head())
# print(d_test.head())

# 데이터 정보 보기
# print(d_train.info())

# 결측치 보기
# print(d_train.isnull().sum()) # 없음
# print(d_test.isnull().sum()) # 없음

# target 빈도
# print(d_train['Outcome'].value_counts()) # 0: 403 / 1: 211

# 타겟 변수 설정
target = d_train.pop('Outcome')

# 검증 데이터 분할
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    d_train, target,
    test_size=0.2, random_state=42)

# 분할된 데이터 크기
# print(X_train.shape, X_val.shape, y_train.shape, y_val.shape) # (491, 8) (123, 8) (491,) (123,)

# 머신러닝 학습 및 평가
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
pred = rf.predict_proba(X_val) # 확률로 예측

# print(f"예측 결과 확인: {pred[:5]}")

# 모델 평가
from sklearn.metrics import roc_auc_score

roc_auc = roc_auc_score(y_val, pred[:,1])
# print(f"roc auc score: {round(roc_auc,3)*100}%") # 82.6%

# 예측 및 결과 파일 생성
pred = rf.predict_proba(d_test)
submit = pd.DataFrame({'pred': pred[:, 1]})
# print(submit)

submit.to_csv('d_result.csv', index=False)

# 제출 파일 확인
print(pd.read_csv('d_result.csv').head())
```

## 이진분류 연습2: 이직 여부 예측
1. 새로운 일자리를 찾을지 예측
2. 예측할 컬럼: target(0: 새 일자리를 찾지 않음. 1: 새 일자리를 찾음)
3. 제출 파일명: h_result.csv (pred: 예측값(이직할 확률) 컬럼 포함)
4. 제출 모델의 성능은 ROC-AUC 평가지표에 따라 체점

```python
hr_train = pt2_ch6 + "/hr_train.csv"
hr_test = pt2_ch6 + "/hr_test.csv"
```

```python
# 데이터 로드
import pandas as pd

h_train = pd.read_csv(hr_train)
h_test = pd.read_csv(hr_test)

# 데이터 크기
# print(h_train.shape, h_test.shape) # (15326, 14) (3832, 13)

# 데이터 샘플
# print(h_train.head())
# print(h_test.head())

# 데이터 정보
# print(h_train.info())
# print(h_test.info())

# 결측치 정보
# print(h_train.isnull().sum())
# print(h_test.isnull().sum())

# 카테고리 수
# print(h_train.nunique())
# print(h_test.nunique())

# target 빈도
# print(h_train['target'].value_counts()) # 0: 11517, 1: 3809

# target 변수 설정
target = h_train.pop('target')

# 결측치 채우기 ('X'로 채움)
h_train = h_train.fillna('X')
h_test = h_test.fillna('X')
# print(h_train.isnull().sum())
# print(h_test.isnull().sum())

# 원-핫 인코딩
combined = pd.concat([h_train, h_test])
combined_dummies = pd.get_dummies(combined)
n_train = len(h_train)
h_train = combined_dummies[:n_train]
h_test = combined_dummies[n_train:]
# print(h_train.head())

# 검증 데이터 분할
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    h_train,
    target,
    test_size = 0.2,
    random_state=42)
# print(X_train.shape, X_val.shape, y_train.shape, y_val.shape) # (12260, 195) (3066, 195) (12260,) (3066,)

# 모델 학습 및 평가
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
pred = rf.predict_proba(X_val)

from sklearn.metrics import roc_auc_score
roc_auc = roc_auc_score(y_val, pred[:, 1])
# print(f"roc auc score: {round(roc_auc, 3)}") # 0.798

# 예측 결과 및 파일 생성
pred = rf.predict_proba(h_test)
submit = pd.DataFrame({'pred': pred[:, 1]})
submit.to_csv('h_result.csv', index=False)
print(pd.read_csv('h_result.csv').head())
```

## 이진분류 연습3: 신용카드 신청자의 미래 신용 예측
1. 신용카드 신청자의 채무 불이행 예측
2. 예측할 컬럼: STATUS(0: 채무 이행, 1: 채무 불이행)
3. 제출 파일명: c_result.csv (pred: 예측값 컬럼 포함)
4. 모델의 성능은 f1-score 로 평가

```python
creditcard_train = pt2_ch6 + '/creditcard_train.csv'
creditcard_test = pt2_ch6 + '/creditcard_test.csv'
```

### 1차: 원-핫 인코딩, 결측치 ('X') 처리
- 데이터불균형으로 f1-score: 0.244

```python
import pandas as pd

# 데이터 불러오기
c_train = pd.read_csv(creditcard_train)
c_test = pd.read_csv(creditcard_test)

# 데이터 크기
# print(c_train.shape, c_test.shape) # (25519, 19) (7591, 18)

# 데이터 샘플
# print(c_train.head())
# print(c_test.head())

# 데이터 정보
# print(c_train.info())
# print(c_test.info())

# 결측치 확인
# print(c_train.isnull().sum())
# print(c_test.isnull().sum()) # 결측치 없음
# print(c_test.describe())

# 결측치 채우기 (train)
c_train = c_train.fillna('X')
# print(c_train.isnull().sum())

# print(c_train['STATUS'].value_counts()) # 0: 25085, 1: 434 > 데이터 불균형 심함

# target 변수 설정
target = c_train.pop('STATUS')

# 원-핫 인코딩
combined = pd.concat([c_train, c_test])
combined_dummies = pd.get_dummies(combined)
n_train = len(c_train)
c_train = combined_dummies[:n_train]
c_test = combined_dummies[n_train:]
# print(c_train.shape, c_test.shape) # (25519, 56) (7591, 56)

# 검증 데이터 분할
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    c_train,
    target,
    test_size = 0.2,
    random_state = 42)

# 머신러닝 학습 및 평가
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
pred = rf.predict(X_val)

from sklearn.metrics import f1_score

score = f1_score(y_val, pred)
# print(f" f1-score: {round(score, 3)}") # 0.244 : f1_score는 1에 가까울수록 좋음

```

### 2차: 결측치 최빈값 처리, 레이블 인코딩

```python
import pandas as pd

# 데이터 불러오기
c_train = pd.read_csv(creditcard_train)
c_test = pd.read_csv(creditcard_test)

# 데이터 크기
# print(c_train.shape, c_test.shape) # (25519, 19) (7591, 18)

# 데이터 샘플
# print(c_train.head())
# print(c_test.head())

# 데이터 정보
# print(c_train.info())
# print(c_test.info())

# 결측치 확인
# print(c_train.isnull().sum())
# print(c_test.isnull().sum()) # 결측치 없음
# print(c_test.describe())

# 결측치 채우기 (train) - 최빈값
freq = c_train['OCCUPATION_TYPE'].mode()[0]
c_train = c_train.fillna(freq)

# # target 변수 설정
target = c_train.pop('STATUS')

# 레이블 인코딩
from sklearn.preprocessing import LabelEncoder

cols = c_train.select_dtypes(include='object').columns
for col in cols:
  le = LabelEncoder()
  c_train[col] = le.fit_transform(c_train[col])
  c_test[col] = le.fit_transform(c_test[col])

# # 검증 데이터 분할
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    c_train,
    target,
    test_size=0.2,
    random_state=42)

# print(X_train.shape, X_val.shape, y_train.shape, y_val.shape) # (20415, 18) (5104, 18) (20415,) (5104,)

# 모델 학습 및 훈련
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
pred = rf.predict(X_val)

from sklearn.metrics import f1_score

score = f1_score(y_val, pred)
# print(f"f1-score: {round(score, 3)}") # 0.277

# 에측 및 결과 파일 생성
pred = rf.predict(c_test)
submit = pd.DataFrame({'pred': pred})
submit.to_csv('c_result.csv', index=False)

print(pd.read_csv('c_result.csv').head())
```

```python
import pandas as pd

score_train = pd.read_csv(credit_score_train)
score_test = pd.read_csv(credit_score_test)

# 데이터 크기 확인
# print(score_train.shape, score_test.shape) # (4198, 21) (1499, 20)

# train 데이터 샘플 확인
# print(score_train.head())

# 데이터 정보 확인
# print(score_train.info())
# print(score_test.info())

# 통계 정보 확인
# print(score_train.describe())

# 결측치 확인
# print(score_train.isnull().sum()) # 없음
# print(score_test.isnull().sum()) # 없음

# target 변수 설정
target = score_train.pop('Credit_Score')
# target 빈도 확인
# print(target.value_counts()) # Standard: 2225 / Poor: 1232 /  Good: 741

# 원-핫 인코딩
combined_df = pd.concat([score_train, score_test])
combined_dummies = pd.get_dummies(combined_df)
n_train = len(score_train)
score_train = combined_dummies[:n_train]
score_test = combined_dummies[n_train:]
# print(score_train.shape, score_test.shape) # (4198, 29) (1499, 29)

# 검증 데이터 분할
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    score_train,
    target,
    test_size = 0.2,
    random_state=42)

# print(X_train.shape, X_val.shape, y_train.shape, y_val.shape) # (3358, 29) (840, 29) (3358,) (840,)

# 머신러닝 학습
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
pred = rf.predict(X_val)

# 모델 평가
from sklearn.metrics import f1_score

score = f1_score(y_val, pred, average='macro')
# print(f"F1-Score: {round(score, 3)}")

# 예측 및 제출
pred = rf.predict(score_test)
submit = pd.DataFrame({'pred': pred})
submit.to_csv('cr_result.csv', index=False)

# 확인
result = pd.read_csv('cr_result.csv')
print(result.shape)
print(result.head())
```
