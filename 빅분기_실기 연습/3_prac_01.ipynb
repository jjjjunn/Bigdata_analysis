# 단일 표본 검정
문제. 커피 제조회사에서 새로 출시한 커피의 카페인 함량이 평균 95mg 미만이라는 주장이 타당한지를 유의수준 5%에서 검정

1. 표본 데이터 평균 구하기
2. Shapiro-Wilk 검정의 p-value 구하기
3. 단일 표본 t-검정의 검정 통계량 구하기
4. 단일 표본 t-검정의 p-value 구하기
5. 유의수준 0.05 하에서 귀무가설을 기준으로 검정 결과를 채택/기각 중 선택해 출력

```python
import pandas as pd
df = pd.DataFrame({
    'Caffeine(mg)': [
        94.2, 93.7, 95.5, 93.9, 94.0, 95.2, 94.7, 93.5, 92.8, 94.4,
        93.8, 94.6, 93.3, 95.1, 94.3, 94.9, 93.9, 94.8, 95.0, 94.2,
        93.7, 94.4, 95.1, 94.0, 93.6
    ]
})

# print(df)

# 1. 표본 데이터 평균 구하기
mean = df['Caffeine(mg)'].mean()
print(f"표본평균: {mean}") # 94.264

# 2. Shapiro-Wilk 검정의 p-value 구하기
from scipy import stats

# print(help(stats))
# print(dir(stats))
# print(help(stats.shapiro))
# shapiro_result = stats.shapiro(df)
# print(shapiro_result)
sh_stat, sh_p_value = stats.shapiro(df)
print(f"shapiro p-value: {sh_p_value}") # 0.9322

# 3. 단일 표본 t-검정의 검정 통계량 구하기
# print(help(stats.ttest_1samp))

coffee_mean = 95 # 회사가 주장하는 카페인 평균 수치
# result = stats.ttest_1samp(df, coffee_mean, alternative='less') # less: 귀무가설 평균보다 작다
stat, pvalue = stats.ttest_1samp(df['Caffeine(mg)'], coffee_mean, alternative='less')
# print(result[0])
print(f"검정통계량: {stat}")

# 4. 단일 표본 t-검정의 p-value 구하기
# print("{:.10f}".format(result[1]))
# print(pvalue)
print(f"p-value: {pvalue:.10f}") # 포맷팅을 위해 ttest에서 df[컬럼명] 형태로 인자 입력

# 5. 유의수준 0.05 하에서 귀무가설을 기준으로 검정 결과를 채택/기각 중 선택해 출력
if pvalue <= 0.05:
  print("귀무가설 기각")
else:
  print("귀무가설 채택")
```

# 독립 표본 검정
문제. 새로 개발한 스마트폰 충전기의 효과를 검정하기 위해 사용자 집단을 두 그룹으로 나눠 두 그룹의 평균 충전 완료 시간 비교
그룹1: 새로운 충전기 제공
그룹2: 기존 충전기 제공
> 새로운 충전기가 더 빠르다고 할 수 있는지를 유의수준 5% 하에서 검정

1. 위의 가설을 검정하기 위한 검정 통계량 구하기
2. 위의 통계량에 대한 p-value 구하기
3. 유의수준 0.05 하에서 귀무가설을 기준으로 검정의 결과를 채택/기각 중 선택해 출력

```python
import pandas as pd

df = pd.DataFrame({
    '충전기': ['New'] * 10 + ['Old'] * 10,
    '충전시간': [
        1.5, 1.6, 1.4, 1.7, 1.5, 1.6, 1.7, 1.4, 1.6, 1.5,
        1.7, 1.8, 1.7, 1.9, 1.8, 1.7, 1.8, 1.9, 1.7, 1.6
    ]
})

# print(df)

# 1. 위의 가설을 검정하기 위한 검정 통계량 구하기
from scipy.stats import ttest_ind

# print(dir(ttest_ind))
# print(help(ttest_ind))
new_cond = df['충전기'] == 'New'
old_cond = df['충전기'] == 'Old'

group1 = df[new_cond]
group2 = df[old_cond]

# print(group1.shape, group2.shape)

t_stat, p_value = stats.ttest_ind(group1['충전시간'], group2['충전시간'], alternative='less')
# print(result)
print(f"검정통계량: {t_stat}")

# 2. 위의 통계량에 대한 p-value 구하기
print(f"p-value: {p_value:.10f}")

# 3. 유의수준 0.05 하에서 귀무가설을 기준으로 검정의 결과를 채택/기각 중 선택해 출력
if p_value <= 0.05:
  print("p-value가 유의수준보다 작으므로, 귀무가설 기각")
else:
  print("p-value가 유의수준보다 크므로, 귀무가설 채택")
```

# 대응 표본 검정
문제. 기존의 교육 방법과 새로운 교육 방벙을 적용한 결과에 대해 유의수준 5%에서 새로운 교육 방법이 교육 시간을 단축시켰는지 검정 (모집단은 정규 분포 가정)

1. 새로움 - 기존의 평균의 표본 평균 구하기
2. 위의 가설을 검정하기 위한 검정 통계량 구하기
3. 위의 통계량에 대한 p-value 구하기
4. 유의수준 0.05 하에서 귀무가설을 기준으로 검정의 결과를 채택/기각 중 선택해 출력

```python
import pandas as pd
df = pd.DataFrame({
    'User': list(range(1, 11)),
    '기존방법': [60.4, 60.7, 60.5, 60.3, 60.8, 60.6, 60.2, 60.5, 60.7, 60.4],
    '새로운방법': [59.8, 60.2, 60.1, 59.9, 59.7, 58.4, 57.0, 60.3, 59.6, 59.8]
})

# print(df)

# 1. 새로움 - 기존의 평균의 표본 평균 구하기
mean_new = df['새로운방법'].mean()
mean_old = df['기존방법'].mean()
mean_diff = mean_new - mean_old
print(f"표본 평균: {mean_diff}")

# 2. 위의 가설을 검정하기 위한 검정 통계량 구하기
from scipy import stats
stat, p_value = stats.ttest_rel(df['새로운방법'], df['기존방법'], alternative='less') # 새로운 것이 교육시간을 단축시켰으므로 less
print(f"검정통계량: {stat}")

# 3. 위의 통계량에 대한 p-value 구하기
print(f"p-value: {p_value:.10f}")

# 4. 유의수준 0.05 하에서 귀무가설을 기준으로 검정의 결과를 채택/기각 중 선택해 출력
if p_value <= 0.05:
  print("p-value가 유의수준보다 작으므로, 귀무가설 기각")
else:
  print("p-value가 유의수준보다 크므로, 귀무가설 채택")
```

# 일원 분산 분석
문제. 4개의 중학교에서 다른 교육 방법을 사용해 학생들에게 수학을 가르쳤는데, 4개의 교육 방법에 따른 성적 차이가 통게적으로 유의미한지 5%의 유의수준에서 검정하라.

1. 각 그룹의 수학 성적에 대해 Shapiro-Wilk 검정을 통한 정규성 확인하고, p-value 구하기
2. 4개 그룹의 수학 성적이 등분산성을 갖는지 확인하기 위해 Levene 검정 실시하고, p-value 구하기
3. 유의수준 0.05 하에서 귀무가설을 기준으로 검정의 결과를 채택/기각 중 선택해 출력
4. 그룹 변수의 자유도 출력
5. 잔차의 자유도 출력
6. 성적의 제곱합 출력
7. 성적의 평균 제곱 출력
8. F-통계량의 값 출력
9. 성적에 대한 p-value 출력

```python
import pandas as pd
# df = pd.read_csv("math.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part3/ch6/math.csv")
# print(df.head())

from scipy import stats

# 1. 각 그룹의 수학 성적에 대해 Shapiro-Wilk 검정을 통한 정규성 확인하고, p-value 구하기
cond_a = df['groups'] == 'group_A'
cond_b = df['groups'] == 'group_B'
cond_c = df['groups'] == 'group_C'
cond_d = df['groups'] == 'group_D'

group_a = df[cond_a]['scores']
group_b = df[cond_b]['scores']
group_c = df[cond_c]['scores']
group_d = df[cond_d]['scores']

group_list = [group_a, group_b, group_c, group_d]

for group in group_list:
  sh_stat, sh_p = stats.shapiro(group)
  print(f"샤피로 p-value: {sh_p}")

# 2. 4개 그룹의 수학 성적이 등분산성을 갖는지 확인하기 위해 Levene 검정 실시하고, p-value 구하기
le_stat, le_p = stats.levene(group_a, group_b, group_c, group_d)
print(f"Levene p-value: {le_p}")

# 3. 유의수준 0.05 하에서 귀무가설을 기준으로 검정의 결과를 채택/기각 중 선택해 출력
# import statsmodels
# print(dir(statsmodels.formula.api))
# # print(dir(statsmodels.formula.api))
from statsmodels.formula.api import ols

model = ols('scores ~ groups', df).fit()

from statsmodels.stats.anova import anova_lm
# print(dir(statsmodels.stats.anova))

print(anova_lm(model))

# 4. 그룹 변수의 자유도 출력
print(f"그룹 자유도: {anova_lm(model).iloc[0,0]}")

# 5. 잔차의 자유도 출력
print(f"잔차 자유도: {anova_lm(model).iloc[1,0]}")

# 6. 성적의 제곱합 출력
print(f"성적 제곱합: {anova_lm(model).iloc[0,1]}")

# 7. 성적의 평균 제곱 출력
print(f"성적 평균 제곱: {anova_lm(model).iloc[0,2]}")

# 8. F-통계량의 값 출력
print(f"F-통계량: {anova_lm(model).iloc[0,3]}")

# 9. 성적에 대한 p-value 출력
score_p = anova_lm(model).iloc[0,4]
print(f"성적 p-value: {score_p:.10f}")
if score_p <= 0.05: # p값이 유의수준 보다 작으면
  print("p-value가 유의수준보다 작으므로 귀무가설 기각")
else:
  print("p-value가 유의수준보다 크므로 귀무가설 채택")
```

# 이원 분산 분석
문제. 12개의 토마토 식물을 무작위로 선택해 세 가지 다른 비료 유형과 네 가지 다른 물주기 실험을 통해 토마토 수확량 증가 연구

1. 비료 유형에 따른 토마토 수확량의 평균 차이가 있는지 검정하기 위한 검정 통계량 구하기
2. 위의 통계량에 대한 p-value 구하기
3. 위의 검정 결과를 유의수준 0.05 하에서 귀무가설을 기준으로 채택/기각 중 선택해 출력
4. 물 주기에 따른 토마토 수확량의 평균에 차이가 있는지 검정하기 위한 검정 통계량 구하기
5. 위의 통계량에 대한 p-value 구하기
6. 위의 검정 결과를 유의수준 0.05 하에서 귀무가설을 기준으로 채택/기각 중 선택해 출력
7. 비료 유형과 물 주기 간의 상호작용이 토마토 수확량에 영향이 있는지 검정하기 위한 검정 통계량 구하기
8. 위의 통계량에 대한 p-value 구하기
9. 위의 검정 결과를 유의수준 0.05 하에서 귀무가설을 기준으로 채택/기각 중 선택해 출력

```python
import pandas as pd
# df = pd.read_csv("tomato2.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part3/ch6/tomato2.csv")
print(df.head())

# 1. 비료 유형에 따른 토마토 수확량의 평균 차이가 있는지 검정하기 위한 검정 통계량 구하기
from statsmodels.formula.api import ols

model = ols('수확량 ~ C(비료유형) + C(물주기) + C(비료유형):C(물주기)', df).fit()

from statsmodels.stats.anova import anova_lm

anova_table = anova_lm(model)
print(anova_table)

feed_f = anova_table.iloc[0, 3]
print(f"\n비료 유형 검정 통계량: {feed_f}")

# 2. 위의 통계량에 대한 p-value 구하기


feed_p = anova_table.iloc[0, 4]
print(f"\n비료유형 p-값: {feed_p:.10f}")

# 3. 위의 검정 결과를 유의수준 0.05 하에서 귀무가설을 기준으로 채택/기각 중 선택해 출력

if feed_p <= 0.05: # 비료유형 p값이 유의수준보다 작으면
  print("p-value가 유의수준보다 작으므로, 귀무가설 기각")
else:
  print("p-value가 유의수준보다 크므로, 귀무가설 채택")

# 4. 물 주기에 따른 토마토 수확량의 평균에 차이가 있는지 검정하기 위한 검정 통계량 구하기
water_f = anova_table.iloc[1, 3]
print(f"\n물 주기 검정 통계량: {water_f}")

# 5. 위의 통계량에 대한 p-value 구하기
water_p = anova_table.iloc[1, 4]
print(f"물 주기 p-value: {water_p}")

# 6. 위의 검정 결과를 유의수준 0.05 하에서 귀무가설을 기준으로 채택/기각 중 선택해 출력
if water_p <= 0.05: # p값이 유의수준보다 작으면
  print("p값이 유의수준보다 작으므로, 귀무가설 기각")
else:
  print("p값이 유의수준보다 크므로, 귀무가설 채택")

# 7. 비료 유형과 물 주기 간의 상호작용이 토마토 수확량에 영향이 있는지 검정하기 위한 검정 통계량 구하기
wa_feed_f = anova_table.iloc[2, 3]
print(f"\n상호작용 검정 통계량: {wa_feed_f}")

# 8. 위의 통계량에 대한 p-value 구하기
wa_feed_p = anova_table.iloc[2, 4]
print(f"\n상호작용 p값: {wa_feed_p}")

# 9. 위의 검정 결과를 유의수준 0.05 하에서 귀무가설을 기준으로 채택/기각 중 선택해 출력
if wa_feed_p <= 0.05:
  print("p값이 유의수준보다 작으므로, 귀무가설 기각")
else:
  print("p값이 유의수준보다 크므로, 귀무가설 채택")
```

# 적합도 검정
문제. 특정 도시 운전자들의 교통사고 경험 수 분포가 전국적인 영향을 따르는지 유의수준 0.05 하에서 검정

1. 이 도시의 교통사고 경험자 중 5회 이상의 비율을 0과 1 사이로 구하기
2. 이 도시 운전자들의 교통사고 경험 수 분포가 전국적인 경향을 따르는지를 검정하기 위한 검정 통계량 구하기
3. 위의 통계량에 대한 p-value 구하기
4. 위의 검정 결과를 유의수준 0.05 하에서 귀무가설을 기준으로 채택/기각 중 선택해 출력

```python
import pandas as pd

# 교통사고 경험 수
df = pd.DataFrame({
    '구분': ['1회', '2회', '3회', '4회', '5회'],
    '경험수': [550, 250, 100, 70, 30],
    '전국적인경향': [0.6, 0.25, 0.08, 0.05, 0.02]})

# print(df)

total_people = df['경험수'].sum()
# print(total_people)

# 1. 이 도시의 교통사고 경험자 중 5회 이상의 비율을 0과 1 사이로 구하기
over_5 = df.iloc[4, 1]
# print(over_5)
print(f"1. 5회 이상 교통사고 경험 비율: {over_5/total_people}")

# 2. 이 도시 운전자들의 교통사고 경험 수 분포가 전국적인 경향을 따르는지를 검정하기 위한 검정 통계량 구하기
# import scipy.stats
# print(dir(scipy.stats))
from scipy.stats import chisquare

observed = [550, 250, 100, 70, 30]
expected = [1000 * 0.6, 1000*0.25, 1000*0.08, 1000*0.05, 1000*0.02]

static, p_value = chisquare(observed, expected)
print(f"2. 검정통계량: {static}")

# 3. 위의 통계량에 대한 p-value 구하기
print(f"3. p-value: {p_value:.10f}")

# 4. 위의 검정 결과를 유의수준 0.05 하에서 귀무가설을 기준으로 채택/기각 중 선택해 출력
if p_value <= 0.05:
  print("p-value가 유의수준보다 작으므로, 귀무가설 기각")
else:
  print("p-value가 유의수준보다 크므로, 귀무가설 채택")
```

# 독립성 검정

문제. 대학교 캠프에서 주최하는 세미나 등록 여부를 조사하여 캠프와 세미나 등록 여부가 서로 독립적인지 유의수준 0.05 하에서 검정

1. 캠프와 세미나 등록 여부가 서로 독립적인지를 검정하기 위한 검정 통계량 구하기
2. 위의 통계량에 대한 p-value 구하기
3. 위의 검정 결과를 유의수준 0.05 하에서 귀무가설을 기준으로 채택/기각 중 선택해 출력

```python
import pandas as pd

df = pd.DataFrame({
    '구분': ['빅데이터분석', '정보처리기사'],
    '등록': [50, 60],
    '미등록': [30, 40]})

print(df)

# 1. 캠프와 세미나 등록 여부가 서로 독립적인지를 검정하기 위한 검정 통계량 구하기
# import scipy.stats
# print(dir(scipy.stats))
from scipy.stats import chi2_contingency

# print(help(chi2_contingency))

obs = pd.DataFrame([[50, 60], [30, 40]])

result = chi2_contingency(obs)
# print(result)
print(f"1. 검정통계량: {result.statistic}")

# 2. 위의 통계량에 대한 p-value 구하기
print(f"2. p-value: {result. pvalue}")

# 3. 위의 검정 결과를 유의수준 0.05 하에서 귀무가설을 기준으로 채택/기각 중 선택해 출력
print("p-value가")
if result.pvalue <= 0.05:
  print("유의수준보다 작으므로 귀무가설 기각")
else:
  print("유의수준보다 크므로 귀무가설 채택")
```

# 다중 선형 회귀
문제. 배달 서비스의 아래 요인들을 이용해 주문량을 예측하는 다중 선형 회귀 모델을 구축하려고 한다.
> 요인(독립변수): 할인율, 온도, 광고비

1. 할인율과 온도의 상관 계수 구하기 (반올림하여 소수 둘째 자리까지)
2. 모델의 결정 계수 구하기 (반올림하여 소수 둘째 자리까지)
3. 각 변수의 회귀 계수 구하기 (반올림하여 소수 넷째 자리까지)
4. 모델의 절편(intercept) 구하기 (반올림하여 소수 넷째 자리까지)
5. 온도의 회귀 계수가 통계적으로 유의한지 검정 (p-value를 반올림하여 소수 넷째 자리까지)
6. 데이터가 할인율이 10%이고, 온도가 20도, 광고비가 500,000일 때 배달 주문량 예측
7. 잔차 제곱합 구하기
8. MSE 구하기
9. 온도의 회귀 계수에 대한 90% 신뢰 구간 구하기
10. 데이터가 할인율이 15%, 온도가 25도, 광고비가 300,000일 때 90% 신뢰 구간과 예측 구간 구하기
11. 독립변수 할인율과 온도를 고정한 상태에서 배달비에 영향을 주는지 가설 검정하고, 유의 수준 0.05를 기준으로 귀무가설 기각 또는 채택으로 출력

```python
import pandas as pd
df = pd.DataFrame({
    '할인율': [28, 24, 13, 0, 27, 30, 10, 16, 6, 5, 7, 11, 11, 30, 25,
            4, 7, 24, 19, 21, 6, 10, 26, 13, 15, 6, 12, 6, 20, 2],
    '온도': [15, 34, 15, 22, 29, 30, 14, 17, 28, 29, 19, 19, 34, 10,
           29, 28, 12, 25, 32, 28, 22, 16, 30, 11, 16, 18, 16, 33, 12, 22],
    '광고비': [342, 666, 224, 764, 148, 499, 711, 596, 797, 484, 986, 347, 146, 362, 642,
            591, 846, 260, 560, 941, 469, 309, 730, 305, 892, 147, 887, 526, 525, 884],
    '주문량': [635, 958, 525, 25, 607, 872, 858, 732, 1082, 863, 904, 686, 699, 615, 893,
            830, 856, 679, 918, 951, 789, 583, 988, 631, 866, 549, 910, 946, 647, 943]
})
# print(df.head())

from statsmodels.formula.api import ols

model = ols('주문량 ~ 할인율 + 온도 + 광고비', df).fit()
# print(model.summary())

# 1. 할인율과 온도의 상관 계수 구하기 (반올림하여 소수 둘째 자리까지)
dis_temp_corr = df['할인율'].corr(df['온도'])
print(f"1. 할인율과 온도의 상관계수: {round(dis_temp_corr, 2)}")

# 2. 모델의 결정 계수 구하기 (반올림하여 소수 둘째 자리까지)
# print(dir(model))
print(f"\n2. 모델 결정 계수: {round(model.rsquared, 2)}")

# 3. 각 변수의 회귀 계수 구하기 (반올림하여 소수 넷째 자리까지)
print(f"\n3. 각 변수 회귀 계수: {round(model.params, 4)}")

# 4. 모델의 절편(intercept) 구하기 (반올림하여 소수 넷째 자리까지)
print(f"\n4. 모델의 절편: {round(model.params['Intercept'], 4)}")

# 5. 온도의 회귀 계수가 통계적으로 유의한지 검정 (p-value를 반올림하여 소수 넷째 자리까지)
print(f"\n5. 온도 p-value: {round(model.pvalues['온도'], 4)}")

# 6. 데이터가 할인율이 10%이고, 온도가 20도, 광고비가 500,000일 때 배달 주문량 예측
new_data = pd.DataFrame({
    '할인율': [10],
    '온도': [20],
    '광고비': [500]})

result = model.predict(new_data)
print(f"\n6. 예측 주문량: {result}")

# 7. 잔차 제곱합 구하기
df['잔차'] = df['주문량'] - model.predict(df)
print(f"\n7. 잔차 제곱합: {sum(df['잔차'] **2)}")

# 8. MSE 구하기
mse = (df['잔차'] **2).mean()
print(f"\n8. MSE: {mse}")

# 9. 온도의 회귀 계수에 대한 90% 신뢰 구간 구하기
conf_90 = model.conf_int(alpha=0.1)
print(f"\n9. 90% 신뢰 구간: {conf_90.loc['온도']}")

# 10. 데이터가 할인율이 15%, 온도가 25도, 광고비가 300,000일 때 90% 신뢰 구간과 예측 구간 구하기
new_data2 = pd.DataFrame({
    '할인율': [15],
    '온도': [25],
    '광고비': [300]})

pred = model.get_prediction(new_data2)
result = pred.summary_frame(alpha=0.1)
print(f"\n10. 90% 신뢰구간과 예측 구간: {result}")

# 11. 독립변수 할인율과 온도를 고정한 상태에서 배달비에 영향을 주는지 가설 검정하고, 유의 수준 0.05를 기준으로 귀무가설 기각 또는 채택으로 출력
cond = model.pvalues['광고비'] < 0.05
if cond:
  opinion = "기각"
else:
  opinion = "채택"
print(f"\n11.귀무가설: {opinion}")
```

# 로지스틱 회귀

문제. 주어진 데이터를 사용하여 데이터를 앞 50%의 a와 나머지 50%의 b로 나누고,
a 데이터로 로지스틱 회귀 모델을 적합

1. 종속변수 target과 모든 독립변수를 사용하여 로지스틱 회귀 모델을 적합하고,
유의하지 않은 독립변수의 개수 구하기 (유의수준 0.05)
2. p-value가 0.05보다 작은 유의한 변수만 사용하여 수정된 모델을 만들고 적합.
이 쉉된 모델에서 가장 큰 p-value를 가진 변수의 이름 구하기
3. 수정된 모델에서 독립변수 중 절댓값이 가장 큰 회귀계수를 가진 변수의 이름 구하기
4. 수정된 모델에서 로그 우도 구하기
5. 수정된 모델에서 잔차 이탈도 구하기
6. 수정된 모델에서 'booked' 변수가 3 증가할 때 오즈비 계산
7. 수정된 모델에서 p-value 가 0.05 보다 작은 회귀계수의 총합 계산.
상수항(절편) 회귀계수도 유의할 경우 포함
8. 수정된 모델로 b 데이터를 사용해 에측한 후, b 데이터의 target과 비교해 정확도 계산 (0과 1사이의 값)
9. 8에서 계산한 정확도를 바탕으로 오류율 계산

```python
import pandas as pd
# df = pd.read_csv("customer_travel.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/refs/heads/main/part3/ch6/customer_travel.csv")

# df.head()

# print(df.shape) # (800, 5)

# print(df.info())

# 데이터 분할하기 5:5
n_data = int(len(df) / 2)
a = df[:n_data]
b = df[n_data:]
# print(a.shape, b.shape) # (400, 5) (400, 5)

from statsmodels.formula.api import logit
# 1. 종속변수 target과 모든 독립변수를 사용하여 로지스틱 회귀 모델을 적합하고,
# 유의하지 않은 독립변수의 개수 구하기 (유의수준 0.05)
model = logit('target ~ age + service + social + booked', a).fit()
# print(model.summary())
non_eff_v = sum(model.pvalues[1:]>=0.05)
print(f"1. 유의하지 않은 독립변수 개수: {non_eff_v}")

# 2. p-value가 0.05보다 작은 유의한 변수만 사용하여 수정된 모델을 만들고 적합.
# 이 수정된 모델에서 가장 큰 p-value를 가진 변수의 이름 구하기
eff_v = model.pvalues[1:] <= 0.05
# print(eff_v) # age, booked

new_model = logit('target ~ age + booked', a).fit()
# print(new_model.summary())
print(f"2. 수정 모델에서 가장 큰 p값 변수: {new_model.pvalues[1:].idxmax()}")

# 3. 수정된 모델에서 독립변수 중 절댓값이 가장 큰 회귀계수를 가진 변수의 이름 구하기
print(f"3. 수정 모델에서 절대값의 회귀계수가 가장 큰 변수: {new_model.params[1:].abs().idxmax()}")

# 4. 수정된 모델에서 로그 우도 구하기
print(f"4. 로그 우도: {new_model.llf}")

# 5. 수정된 모델에서 잔차 이탈도 구하기
print(f"5. 잔차 이탈도: {-2 * new_model.llf}")

# 6. 수정된 모델에서 'booked' 변수가 3 증가할 때 오즈비 계산
import numpy as np
print(f"6. 오즈비: {np.exp(new_model.params['booked'] * 3)}")

# 7. 수정된 모델에서 p-value 가 0.05 보다 작은 회귀계수의 총합 계산.
# 상수항(절편) 회귀계수도 유의할 경우 포함
under_5 = sum(new_model.params[new_model.pvalues <= 0.05])
print(f"7. 0.05보다 작은 회귀계수 총합: {under_5}")

# 8. 수정된 모델로 b 데이터를 사용해 에측한 후, b 데이터의 target과 비교해 정확도 계산 (0과 1사이의 값)
pred = new_model.predict(b)
pred = (pred > 0.5).astype(int)

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(b['target'], pred)
print(f"8. 정확도: {accuracy}")

# 9. 8에서 계산한 정확도를 바탕으로 오류율 계산
print(f"9 오류율: {1 - accuracy}")
```
