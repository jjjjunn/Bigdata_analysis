## 작업형 1유형: 문제2
1) 관광객 비율이 두 번째로 높은 국가의 사업 방문객 수를 a라고 정의

2) 관광이 두 번째로 높은 국가의 공무 방문객 수를 b라고 정의

3) a와 b의 합구하기
- 방문객 합계 = 관광 + 공무 + 사업 + 기타
- 관광객 비율 = 관광 / 방문객 합계

```python
ch812 = "https://github.com/lovedlim/bigdata_analyst_cert_v2/raw/refs/heads/main/part4/ch8/tourist.csv"
```

```python
import pandas as pd

df = pd.read_csv(ch812)
# print(df)

# 1) 관광객 비율이 두 번째로 높은 국가의 사업 방문객 수를 a라고 정의
# - 방문객 합계 = 관광 + 공무 + 사업 + 기타
# - 관광객 비율 = 관광 / 방문객 합계
df['방문객합계'] = df.iloc[:, 1:].sum(axis=1)
# print(df)
df['관광객비율'] = df['관광'] / df['방문객합계']
# print(df)
df_sort = df.sort_values('관광객비율', ascending=False)
# print(df_sort) # 203
a = df_sort.iloc[1, 3]
print(f"a: {a}")


# 2) 관광이 두 번째로 높은 국가의 공무 방문객 수를 b라고 정의
df_sort2 = df.sort_values('관광', ascending=False)
# print(df_sort2) # 238
b = df_sort2.iloc[1, 2]
print(f"b: {b}")

# 3) a와 b의 합구하기
result = a + b
print(result)

```

1) 주어진 데이터에서 'co'와 'nmhc' 컬럼을 각각 MinMax 스케일링

2) 스케일링된 'co', 'nmhc' 컬럼의 표준편차 구하기

3) 'co' 컬럼의 표준편차에서 'nmhc' 컬럼의 표준편차를 뺀 값을 \
소수점 3 자리로 반올림하여 구하기

```python
ch813 = "https://github.com/lovedlim/bigdata_analyst_cert_v2/raw/refs/heads/main/part4/ch8/chem.csv"
```

```python
import pandas as pd

df = pd.read_csv(ch813)
# print(df)

# 1) 주어진 데이터에서 'co'와 'nmhc' 컬럼을 각각 MinMax 스케일링
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
# print(help(scaler))
df['co_scaled'] = scaler.fit_transform(df[['co']])
df['nmhc_scaled'] = scaler.fit_transform(df[['nmhc']])
# print(df)

# 2) 스케일링된 'co', 'nmhc' 컬럼의 표준편차 구하기
co_std = df['co_scaled'].std()
nh_std = df['nmhc_scaled'].std()
# print(co_std, nh_std) # 0.2856516497116944 0.3030617020578397


# 3) 'co' 컬럼의 표준편차에서 'nmhc' 컬럼의 표준편차를 뺀 값을
# 소수점 3 자리로 반올림하여 구하기
result = co_std - nh_std
print(round(result, 3))
```

4) 제출 모델의 성능은 MAE (Mean Absolute Error) 평가지표에 따라 평가

```python
train_8 = "https://github.com/lovedlim/bigdata_analyst_cert_v2/raw/refs/heads/main/part4/ch8/churn_train.csv"
test_8 = "https://github.com/lovedlim/bigdata_analyst_cert_v2/raw/refs/heads/main/part4/ch8/churn_test.csv"
```

```python
import pandas as pd

train = pd.read_csv(train_8)
test = pd.read_csv(test_8)

# print(train.shape, test.shape) # (4116, 19) (1764, 18)
# print(train.head(3)) # target: TotalCharges
# print(test.head(3))

# 타겟 변수 설정
target = train.pop('TotalCharges')

# 데이터 정보 확인
# print(train.info()) # 결측치 없음
# print(test.info()) # 결측치 없음

# 데이터 기술 통계
# print(train.describe())
# print(test.describe())

# object 데이터 unique 확인
# cols = train.select_dtypes(include='O').columns
# print(cols)
# print(train[cols].describe()) # 2부터 4116 까지 - customerID: 4116
# print(test[cols].describe()) # 2부터 4116 까지 - customerID: 4116

# customerID 분리하기
train_id = train.pop('customerID')
test_id = test.pop('customerID')

# 레이블 인코딩
from sklearn.preprocessing import LabelEncoder
cols = train.select_dtypes(include='O').columns

combined = pd.concat([train, test]) # 데이터 병합
le = LabelEncoder()

for col in cols:
  combined[col] = le.fit_transform(combined[col])
  # print(f"{col} 인코딩 완료")

n_train = len(train)
train = combined[:n_train]
test = combined[n_train:]
# print(train.shape, test.shape) # (4116, 17) (1764, 17)
# print(train.head(3)) # 인코딩 완료
# print(test.head(3)) # 인코딩 완료

# 검증 데이터 분할
from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(train, target, test_size=0.2, random_state=42)
# print(X_tr.shape, X_val.shape, y_tr.shape, y_val.shape) # (3292, 17) (824, 17) (3292,) (824,)

# 머신러닝 학습 및 예측
from sklearn.ensemble import RandomForestRegressor

rf= RandomForestRegressor(random_state=42)
rf.fit(X_tr, y_tr)
pred = rf.predict(X_val)
# print(pred[:3])

# 모델 평가
from sklearn.metrics import r2_score, mean_absolute_error

r2 = r2_score(y_val, pred)
mae = mean_absolute_error(y_val, pred)

print(f"R2:{r2}, MAE: {mae}") # R2:0.6020346779073029, MAE: 881.4131559101941

# 최종 예측 및 제출
pred = rf.predict(test)
# print(pred[:3])
submit = pd.DataFrame({'pred': pred})
submit.to_csv('result.csv', index=False)

# 파일 확인
result = pd.read_csv('result.csv')
print(result.shape)
print(result)
```

## 작업형 3유형: 문제1
로지스틱 회귀

1-1. 주어진 데이터에서 로지스틱 회귀 분석을 수행해 유의확률(p-value)이\
0.05 이상인 유의하지 않은 독립변수의 개수 구하기
- 종속변수: Churn, 독립변수: 종속변수를 제외한 모든 변수

1-2. 유의확률이 0.05 미만인 유의한 변수만을 사용해 다시 로지스틱 회귀 분석 수행\
이 회귀식의 유의한 회귀 계수(상수항 포함)의 합계를 구하시오
- 반올림하여 소수 셋째 자리까지 계산

1-3. 문제 1-2에서 수행한 로지스틱 회귀에서 'DataUsage' 변수가 5만큼 증가할 때\
오즈비를 구하시오
- 반올림하여 소수 셋째 자리까지 계산

```python
ch831 = "https://github.com/lovedlim/bigdata_analyst_cert_v2/raw/refs/heads/main/part4/ch8/churn.csv"
```

```python
import pandas as pd

df = pd.read_csv(ch831)
# print(df)

# 1-1. 주어진 데이터에서 로지스틱 회귀 분석을 수행해 유의확률(p-value)이
# 0.05 이상인 유의하지 않은 독립변수의 개수 구하기
# - 종속변수: Churn, 독립변수: 종속변수를 제외한 모든 변수
from statsmodels.formula.api import logit
# print(df.columns) # 'AccountWeeks', 'ContractRenewal', 'DataPlan', 'DataUsage',
       #'CustServCalls', 'DayMins', 'DayCalls', 'MonthlyCharge', 'OverageFee',
       # 'RoamMins'

# cols = df.columns
# print(cols.str.replace("'", "")) # 효과 없음
model = logit('Churn ~ AccountWeeks + ContractRenewal + DataPlan + DataUsage + CustServCalls + DayMins + DayCalls + MonthlyCharge + OverageFee + RoamMins', df).fit()
# print(model.summary())

# 유의확률이 0.05 이상인 독립변수 개수 구하기
cond = model.pvalues >= 0.05
# print(cond)
print(f"유의하지 않은 변수 개수: {sum(cond)}")


# 1-2. 유의확률이 0.05 미만인 유의한 변수만을 사용해 다시 로지스틱 회귀 분석 수행
# 이 회귀식의 유의한 회귀 계수(상수항 포함)의 합계를 구하시오
# - 반올림하여 소수 셋째 자리까지 계산

# 유의한 변수: DataUsage, DayMins
model = logit('Churn ~ DataUsage + DayMins', df).fit()
# print(model.summary())

cond = model.params < 0.05
total_p = sum(model.params)
print(f"유의한 회귀계수 합: {round(total_p, 3)}")

# 1-3. 문제 1-2에서 수행한 로지스틱 회귀에서 'DataUsage' 변수가 5만큼 증가할 때
# 오즈비를 구하시오
# - 반올림하여 소수 셋째 자리까지 계산
import numpy as np

data_5 = np.exp(model.params['DataUsage'] * 5)
print(f"오즈비: {round(data_5, 3)}")
```

## 작업형 3유형; 문제2
다중 선형 회귀

2-1. 주어진 데이터를 이용해 종속변수(PIQ)와 독립변수(Brain, Height, Weight)로\
다중 선형 회귀 분석을 수행.\
이때 유의확률이 가장 작은 변수의 회귀 계수 값 구하기
- 반올림하여 소수 셋째 자리까지 계산

2-2. 문제 2-1에 적합한 모델의 결정 계수(R-squared) 값 구하기
-반올림하여 소수 둘째 자리까지 계산

2-3. 뇌 크기(Brain) 90, 키(Height) 70, 몸무게(Weight) 150일 때의 PIQ 예측
- 반올림하여 정수로 계산

```python
ch832 = "https://github.com/lovedlim/bigdata_analyst_cert_v2/raw/refs/heads/main/part4/ch8/piq.csv"
```

```python
import pandas as pd

df = pd.read_csv(ch832)
# print(df)

# 2-1. 주어진 데이터를 이용해 종속변수(PIQ)와 독립변수(Brain, Height, Weight)로
# 다중 선형 회귀 분석을 수행.
# 이때 유의확률이 가장 작은 변수의 회귀 계수 값 구하기
# - 반올림하여 소수 셋째 자리까지 계산
from statsmodels.formula.api import ols

model = ols('PIQ ~ Brain + Height + Weight', df).fit()
print(model.summary()) # 다중 선형 회귀 분석 수행
# 유의확률이 가장 작은 변수
min_p = model.pvalues.idxmin()
print(min_p) # Brain
# 해당 변수의 회귀 계수
brain_coef = model.params[min_p]
print(f"회귀계수: {round(brain_coef, 3)}") #  2.343

# 2-2. 문제 2-1에 적합한 모델의 결정 계수(R-squared) 값 구하기
# -반올림하여 소수 둘째 자리까지 계산
rsq = model.rsquared
print(f"결정계수: {round(rsq, 2)}") # 0.37

# 2-3. 뇌 크기(Brain) 90, 키(Height) 70, 몸무게(Weight) 150일 때의 PIQ 예측
# - 반올림하여 정수로 계산
new_data = pd.DataFrame({'Brain':[90], 'Height': [70], 'Weight': [150]})
pred = model.predict(new_data)
print(f"예측 PIQ: {int(round(pred))}") # 106
```

```python
import pandas as pd

df = pd.read_csv(ch913)
# print(df)

# 1) 만족도의 결측치는 만족도의 평균값으로 대체
# print(df['만족도'].isnull().sum()) # 120건
s_mean = df['만족도'].mean()
# print(s_mean) # 5.017045454545454
df['만족도'] = df['만족도'].fillna(s_mean)
# print(df['만족도'].isnull().sum()) # 채우기 완료

# 2) 근속연수의 결측치는 같은 부서 내에서 동일한 성과등급을 가진 직원들의
# 평균 근속연수로 대체. 이때, 평균 근속연수는 소수점 이하를 절사하여
# 정수로 변환하여 사용
# print(df['근속연수'].isnull().sum()) # 48건
part_mean = df.groupby(['부서', '성과등급'])['근속연수'].transform('mean').astype(int)
# print(part_mean)
df['근속연수'] = df['근속연수'].fillna(part_mean)
# print(df['근속연수'].isnull().sum()) # 결측치 처리 완료

# 3) 각 직원의 '연봉 / 근속연수' 값을 계산하여, 그 값이 세 번째로 큰 직원의 근속연수 구하기
# print(df)
df['연근'] = df['연봉'] / df['근속연수']
# print(df.sort_values('연근', ascending=False)) # 801 / E0802 / 1
df = df.sort_values('연근', ascending=False)
cond = df['연근'].nlargest(3).index[2]
# print(cond)
third_y = df.loc[cond]['근속연수'] # 1.0
print(f"근속연수: {third_y}")

# 4) 각 직원의 '연봉 / 만족도' 값을 계산하여, 그 값이 두 번째로 큰 직원의
# 교육 참가 횟수 구하기
df['연만'] = df['연봉'] / df['만족도']
df = df.sort_values('연만', ascending=False)
# print(df) # 737 / E0738 / 6
cond2 = df['연만'].nlargest(2).index[1]
# print(cond2)
sec_edu = df.loc[cond2]['교육참가횟수']
print(f"교육 참가횟수: {sec_edu}")

# 5) 3과 4의 값을 더한 결과를 구하기 (정수로 출력)
result = third_y + sec_edu
print(int(result))
```

다중 선형 회귀

1. 주어진 데이터에서 id를 기준으로 다음과 같이 데이터를 분할
- train: 1부터 140

1-1. train 데이터에서 c1, c2, c3, c4, c5를 독립변수로,\

독립변수 중 p-value가 0.05보다 작은 변수의 개수 구하기

1-2. 1-1에서 적합한 회귀모형의 결과를 바탕으로, p-value가 0.05 이하인 유의한\
독립변수만을 선택하여 새로운 다중 회귀모형을 적합한 후, train 데이터에서\
design 의 예측값을 산출하고 예측값과 실제 design 값 사이의 피어슨 상관계수를\
반올림하여 소수 셋째 자리까지 게산

1-3. 1-2에서 적합한 회귀모형을 이용하여 test 데이터에서 design의 에측값을 산출한 후, \
예측값과 실제 design 값 사이의 RMSE(Root Mean Squared Error) 를 반올림하여\
소수 셋째 자리까지 계산

```python
ch931= "https://github.com/lovedlim/bigdata_analyst_cert_v2/raw/refs/heads/main/part4/ch9/design.csv"
```

```python
import pandas as pd

df = pd.read_csv(ch931)
# print(df.shape) # (200, 7)
# print(df)

# 1. 주어진 데이터에서 id를 기준으로 다음과 같이 데이터를 분할
# - train: 1부터 140
# - test: 140부터
train = df[:140]
test = df[140:]
# print(train.shape, test.shape) # (140, 7) (60, 7)

# 1-1. train 데이터에서 c1, c2, c3, c4, c5를 독립변수로,
# design을 종속변수로 하는 다중 회귀모형을 적합한 후, 절편을 제외한
# 독립변수 중 p-value가 0.05보다 작은 변수의 개수 구하기
from statsmodels.formula.api import ols

model = ols('design ~ c1 + c2 + c3 + c4 + c5', train).fit()
# print(model.summary()) # 다중 선형 회귀 적합
cond = model.pvalues[1:] < 0.05
print(f" 작은 변수 개수: {sum(cond)}") # 3개: c1, c2, c4
# print(cond)


# 1-2. 1-1에서 적합한 회귀모형의 결과를 바탕으로, p-value가 0.05 이하인 유의한
# 독립변수만을 선택하여 새로운 다중 회귀모형을 적합한 후, train 데이터에서
# design 의 예측값을 산출하고 예측값과 실제 design 값 사이의 피어슨 상관계수를
# 반올림하여 소수 셋째 자리까지 게산
model = ols('design ~ c1 + c2 + c4', train).fit()
# print(model.summary()) # 새로운 회귀 모형 적합
train['pred'] = model.predict(train)
# print(train)
corr = train['design'].corr(train['pred'])
print(f"상관계수: {round(corr, 3)}") # 0.051


# 1-3. 1-2에서 적합한 회귀모형을 이용하여 test 데이터에서 design의 에측값을 산출한 후,
# 예측값과 실제 design 값 사이의 RMSE(Root Mean Squared Error) 를 반올림하여
# 소수 셋째 자리까지 계산
test['pred'] = model.predict(test)
# print(test)
from sklearn.metrics import root_mean_squared_error

rmse = root_mean_squared_error(test['design'], test['pred'])
print(f"RMSE: {round(rmse, 3)}") # 8.488

```

## 작업형 3유형: 문제2
로지스틱 회귀
- 종속변수: Churn(고객 이탈 여부- 1: 이탈, 0: 유지)
- 독립변수
MonthlyCharges (월 사용 요금)\
CustomerTenure (고객 유지 기간)\
HasPhoneService (전화 서비스 이용 여부)\
HasTechInsurance (기술 보험 가입 여부)

2-1. 주어진 데이터를 사용하여 고객 이탈(Churn)을 예측하는는 로지스틱 회귀 모형을\
적합한 후, MonthlyCharges 변수의 p-value를 반올림하여 소수 셋째 자리까지 구하기

2-2. 적합한 로지스틱 회귀모형을 바탕으로, 전화 서비스를 받지 않은 고객이\
이탈할 확률 대비 전화 서비스를 받은 고객이 이탈할 확률에 대한 오즈비(Odds Ratio)
계산
- 반올림하여 소수 셋째 자리까지 계산

2-3. 적합한 로지스틱 회귀 모형을 이용하여 각 고객의 이탈 확률을 예측한 후,\
예측된 이탈 확률이 0.3 초과하는 고객의 수 구하기

```python
ch932 = "https://github.com/lovedlim/bigdata_analyst_cert_v2/raw/refs/heads/main/part4/ch9/retention.csv"
```

```python
import pandas as pd

df = pd.read_csv(ch932)
# print(df)


# 로지스틱 회귀
# - 종속변수: Churn(고객 이탈 여부- 1: 이탈, 0: 유지)
# - 독립변수
# MonthlyCharges (월 사용 요금)\
# CustomerTenure (고객 유지 기간)\
# HasPhoneService (전화 서비스 이용 여부)\
# HasTechInsurance (기술 보험 가입 여부)
from statsmodels.formula.api import logit

model = logit('Churn ~ MonthlyCharges + CustomerTenure + HasPhoneService + HasTechInsurance', df).fit()
# print(model.summary())


# 2-1. 주어진 데이터를 사용하여 고객 이탈(Churn)을 예측하는 로지스틱 회귀 모형을\
# 적합한 후, MonthlyCharges 변수의 p-value를 반올림하여 소수 셋째 자리까지 구하기
mc_p = model.pvalues['MonthlyCharges']
print(f"p-value: {round(mc_p, 3)}") # 0.008


# 2-2. 적합한 로지스틱 회귀모형을 바탕으로, 전화 서비스를 받지 않은 고객이\
# 이탈할 확률 대비 전화 서비스를 받은 고객이 이탈할 확률에 대한 오즈비(Odds Ratio) 계산
# - 반올림하여 소수 셋째 자리까지 계산
import numpy as np

phone = model.params['HasPhoneService']
# print(phone)
phone_odds = np.exp(phone)
print(f"오즈비: {round(phone_odds, 3)}")

# 2-3. 적합한 로지스틱 회귀 모형을 이용하여 각 고객의 이탈 확률을 예측한 후,\
# 예측된 이탈 확률이 0.3 초과하는 고객의 수 구하기
df['pred'] = model.predict(df)
# print(df)

cond = df['pred'] > 0.3
# print(df[cond])
result = len(df[cond])
print(result) # 65
```

```python
ch913 = "https://github.com/lovedlim/bigdata_analyst_cert_v2/raw/refs/heads/main/part4/ch9/hr.csv"
```

```python
import pandas as pd

df = pd.read_csv(ch913)
# print(df)

# 1) 만족도의 결측치는 만족도의 평균값으로 대체
# print(df['만족도'].isnull().sum()) # 120건
mean_s = df['만족도'].mean()
# print(mean_s) # 5.017
df['만족도'] = df['만족도'].fillna(mean_s)
# print(df['만족도'].isnull().sum()) # 결측치 처리 완료

# 2) 근속연수의 결측치는 같은 부서 내에서 동일한 성과등급을 가진 직원들의\
# 평균 근속연수로 대체. 이때, 평균 근속연수는 소수점 이하를 절사하여 정수로 변환
# print(df['근속연수'].isnull().sum()) # 48건
mean_p = df.groupby(['부서', '성과등급'])['근속연수'].transform('mean').astype(int)
# print(mean_p)
df['근속연수'] = df['근속연수'].fillna(mean_p)

# print(df['근속연수'].isnull().sum()) # 결측치 처리 완료

# 3) 각 직원의 '연봉/근속연수' 값을 계산하여, 그 값이 세 번째로 큰 직원의\
# 근속연수 구하기
df['연근'] = df['연봉'] / df['근속연수']

df = df.sort_values('연근', ascending=False)
# print(df)
third = df['연근'].nlargest(3).index[2] # 801
third_y = df.loc[third]['근속연수']
print(f"근속연수: {third_y}") # 1

# 4) 각 직원의 '연봉/만족도' 값을 계산하여, 그 값이 두 번째로 큰직원의\
# 교육 참가 횟수 구하기
df['연만'] = df['연봉'] / df['만족도']
df = df.sort_values('연만', ascending=False)
# print(df)
sec = df['연만'].nlargest(2).index[1] # 737
# print(sec)
sec_edu = df.loc[sec]['교육참가횟수']
print(f"교육 참가 횟수: {sec_edu}")

# 5) 3번과 4번의 값을 더한 결과 구하기 (정수로 출력)
result = third_y + sec_edu
print(int(result))
```

```python
train_9 = "https://github.com/lovedlim/bigdata_analyst_cert_v2/raw/refs/heads/main/part4/ch9/farm_train.csv"
test_9 = "https://github.com/lovedlim/bigdata_analyst_cert_v2/raw/refs/heads/main/part4/ch9/farm_test.csv"
```

```python
import pandas as pd

train = pd.read_csv(train_9)
test = pd.read_csv(test_9)
# print(train.shape, test.shape) # (4000, 9) (1000, 8)
# print(train.head(3)) # target: 농약검출여부
# print(test.head(3))

# 데이터 정보 확인
# print(train.info()) # 결측치 없음
# print(test.info()) # 결측치 없음

# 타겟 변수 설정
target = train.pop('농약검출여부')

# object 변수 unique 확인
cols = train.select_dtypes(include='O').columns
# print(cols)
# print(train[cols].describe()) # 3부터 15
# print(test[cols].describe()) # 3부터 15

# 레이블 인코딩
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
combined = pd.concat([train, test])
for col in cols:
  combined[col] = le.fit_transform(combined[col])
  # print(f"{col} 인코딩 완료")

n_train = len(train)
train = combined[:n_train]
test = combined[n_train:]
# print(train.shape, test.shape) # (4000, 8) (1000, 8)
# print(train.head(3)) # 인코딩 완료
# print(test.head(3)) # 인코딩 완료

# 검증 데이터 분할
from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(train, target, test_size=0.2, random_state=42)
# print(X_tr.shape, X_val.shape, y_tr.shape, y_val.shape) # (3200, 8) (800, 8) (3200,) (800,)

# 머신러닝 학습 및 예측
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=42)
rf.fit(X_tr, y_tr)
pred = rf.predict(X_val)
# print(pred[:3]) # [0 0 0]

# 모델 평가
from sklearn.metrics import accuracy_score, f1_score

acc = accuracy_score(y_val, pred)
f1 = f1_score(y_val, pred, average='macro')
print(f"정확도: {acc}, F1-Score: {f1}") # 정확도: 0.9625, F1-Score: 0.8680387463168264

# 최종 예측 및 제출
pred = rf.predict(test)
# print(pred[:3])
submit = pd.DataFrame({'pred': pred})
submit.to_csv('result.csv', index=False)

# 파일 확인
result = pd.read_csv('result.csv')
print(result.shape)
print(result)
```

## 작업형 3유형: 문제1
다중 선형 회귀

- 주어진 데이터에서 id를 기준으로 데이터를 분할\
train: 1부터 140\
test : 140 이후\

1-1. train 데이터에서 c1, c2, c3, c4, c5를 독립변수로, design을 종속변수로 하는\
다중 회귀 모형을 적합한 후, 절편을 제외한 독립변수 중 p-value가 0.05보다 작은\
변수의 개수 구하기

1-2. 1-1에서 적합한 회귀모형의 결과를 바탕으로, p-value가 0.05 이하인 유의한\
독립변수만을 선택하여 새로운 다중 회귀모형을 적합한 후, train 데이터에서 design\
예측값을 산출하고 예측값과 실제 design 값 사이의 피어슨 상관계수를 반올림하여\
소수 셋째 자리까지 게산

1-3. 1-2에서 적합한 회귀모형을 이용하여 test 데이터에서 design의 예측값을\
산출한 후, 예측값과 실제 design 값 사이의 RMSE(root mean squared error) 를\
반올림하여 소수 셋째 자리까지 계산

```python
ch931 = "https://github.com/lovedlim/bigdata_analyst_cert_v2/raw/refs/heads/main/part4/ch9/design.csv"
```

```python
import pandas as pd

df = pd.read_csv(ch931)
# print(df)
# print(df.shape) # (200, 7)

# - 주어진 데이터에서 id를 기준으로 데이터를 분할\
# train: 1부터 140\
# test : 140 이후\
train = df[:140]
test = df[140:]
# print(train.shape, test.shape) # (140, 7) (60, 7)

# 1-1. train 데이터에서 c1, c2, c3, c4, c5를 독립변수로, design을 종속변수로 하는\
# 다중 회귀 모형을 적합한 후, 절편을 제외한 독립변수 중 p-value가 0.05보다 작은\
# 변수의 개수 구하기
from statsmodels.formula.api import ols

model = ols('design ~ c1 + c2 + c3 + c4 + c5', train).fit()
# print(model.summary()) # 다중 회귀 모형 적합
cond = model.pvalues[1:] < 0.05
# print(cond) # c1, c2, c4
print(f"유의한 변수 개수: {sum(cond)}")


# 1-2. 1-1에서 적합한 회귀모형의 결과를 바탕으로, p-value가 0.05 이하인 유의한\
# 독립변수만을 선택하여 새로운 다중 회귀모형을 적합한 후, train 데이터에서 design\
# 예측값을 산출하고 예측값과 실제 design 값 사이의 피어슨 상관계수를 반올림하여\
# 소수 셋째 자리까지 게산
model = ols('design ~ c1 + c2 + c4', train).fit()
# print(model.summary()) # 새로운 모형 적합
train['pred'] = model.predict(train) # 새로운 예측값 산출
# print(train)

corr = train['design'].corr(train['pred'])
print(f"상관계수: {round(corr, 3)}") # 0.501

# 1-3. 1-2에서 적합한 회귀모형을 이용하여 test 데이터에서 design의 예측값을\
# 산출한 후, 예측값과 실제 design 값 사이의 RMSE(root mean squared error) 를\
# 반올림하여 소수 셋째 자리까지 계산
test['pred'] = model.predict(test)
# print(test)

from sklearn.metrics import root_mean_squared_error

rmse = root_mean_squared_error(test['design'], test['pred'])
print(f"RMSE: {round(rmse, 3)}") # 8.488
```

## 작업형 3유형: 문제2
로지스틱 회귀

- 종속변수: Churn (고객 이탈 여부- 1:이탈, 0: 유지)
- 독립변수
MonthlyCharges (월 사용 요금)\
CustomerTenure (고객 유지 기간)\
HasPhoneService (전화 서비스 이용 여부)\
HasTechInsurance (기술 보험 가입 여부)

2-1. 주어진 데이터를 사용하여 고객 이탈(Churn)을 예측하는 로지스틱 회귀 모형을\
적합한 후, MonthlyCharges 변수의 p-value를 반올림하여 소수 셋째 자리까지 게산

2-2. 적합한 로지스틱 회귀모형을 바탕으로, 전화 서비스를 받지 않은 고객이\
이탈할 확률 대비 전화 서비스를 받은 고객이 이탈할 확률에 대한 오즈비 계산\
반올림하여 소수 셋째 자리까지 계산

```python
ch932 = "https://github.com/lovedlim/bigdata_analyst_cert_v2/raw/refs/heads/main/part4/ch9/retention.csv"
```

```python
import pandas as pd

df = pd.read_csv(ch932)
# print(df)


# - 종속변수: Churn (고객 이탈 여부- 1:이탈, 0: 유지)
# - 독립변수
# MonthlyCharges (월 사용 요금)\
# CustomerTenure (고객 유지 기간)\
# HasPhoneService (전화 서비스 이용 여부)\
# HasTechInsurance (기술 보험 가입 여부)
from statsmodels.formula.api import logit


# 2-1. 주어진 데이터를 사용하여 고객 이탈(Churn)을 예측하는 로지스틱 회귀 모형을\
# 적합한 후, MonthlyCharges 변수의 p-value를 반올림하여 소수 셋째 자리까지 게산
model = logit('Churn ~ MonthlyCharges + CustomerTenure + HasPhoneService + HasTechInsurance', df).fit()
print(model.summary()) # 0.008
month_p = model.pvalues['MonthlyCharges']
print(f"p-value: {round(month_p, 3)}") # 0.008


# 2-2. 적합한 로지스틱 회귀모형을 바탕으로, 전화 서비스를 받지 않은 고객이\
# 이탈할 확률 대비 전화 서비스를 받은 고객이 이탈할 확률에 대한 오즈비 계산\
# 반올림하여 소수 셋째 자리까지 계산
import numpy as np

phone_param = model.params['HasPhoneService']
# print(phone_param)
phone = np.exp(phone_param)
print(f"오즈비: {round(phone, 3)}") # 0.701
```
