# 회귀 실습

## 라이브러리 및 데이터 로드

```python
import pandas as pd

train = pd.read_csv(train_url)
test = pd.read_csv(test_url)

print(f"훈련 데이터: {train.shape}")
print(f"테스트 데이터: {test.shape}")
```

```python
test['Item_Weight'].fillna(test['Item_Weight'].mean(), inplace=True)
test['Outlet_Size'].fillna(test['Outlet_Size'].mode()[0], inplace=True)

test.isnull().sum()
```

## 데이터 전처리

```python
# obejct 컬럼 선택
cols= list(train.columns[train.dtypes == object])
cols
```

```python
# target 값 따로 저장
target = train.pop('Item_Outlet_Sales')
print(f"데이터 크기: {train.shape}, {test.shape}")

# 인코딩 위한 데이터 병합
df = pd.concat([train, test])
print(f"병합 후 데이터 크기: {df.shape}")
```

## 레이블 인코딩

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for col in cols:
  df[col] = le.fit_transform(df[col])
  print(f"{col} 인코딩 완료")
```

```python
df.head()
```

```python
# 인코딩 후 데이터 분할
train = df.iloc[:len(train)].copy()
test = df.iloc[len(train):].copy()

print("데이터 수")
print(f"train: {train.shape}")
print(f"test: {test.shape}")
```

## 데이터 분할

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    train,
    target,
    test_size=0.2,
    random_state=42)

data_list = [X_train, X_val, y_train, y_val]
for data in data_list:
  print(data.shape)
```

## 머신러닝 학습 및 평가

```python
# 모델 평가 함수
from sklearn.metrics import mean_squared_error, r2_score, root_mean_squared_error, mean_absolute_error

metrics_list = [mean_squared_error, r2_score, root_mean_squared_error, mean_absolute_error]
metrics_names = ['MSE', 'R2', 'RMSE', 'MAE']

def regressor_metrics(y_val, y_pred):
  for metrics, name in zip(metrics_list, metrics_names):
    score = metrics(y_val, y_pred)
    print(f"{name}: {round(score, 3)}")
```

```python
# 선형 회귀
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_val)

regressor_metrics(y_val, y_pred_lr)
```

```python
# 랜덤 포레스트
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_val)

regressor_metrics(y_val, y_pred_rf)
```

```python
# LightGBM
import lightgbm as lgb

model = lgb.LGBMRegressor(random_state=42)
model.fit(X_train, y_train)
y_pred_lgb = model.predict(X_val)

regressor_metrics(y_val, y_pred_lgb)
```
