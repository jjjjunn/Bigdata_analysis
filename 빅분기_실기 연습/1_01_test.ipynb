## 판다스 기초 문제

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "제품명": ['아반떼', '소나타', '클레오스', '말리부', '토레스', '팰리세이드'],
    "가격(만원)": [3000, 3500, 4000, 3800, 3200, 6000],
    "연식": [2024, 2023, np.nan, 2020, np.nan, np.nan],
    "브랜드": ['현대', '현대', '르노', '쉐보레', 'KGM', np.nan],
})

df
```

1. '연식' 컬럼의 결측치를 연식 데이터 중 최솟값으로 채우기
2. '브랜드' 컬럼의 결측치를 브랜드 컬럼의 최빈값으로 채우기
3. 가격이 3500 이상인 데이터의 수 구하기

5. '연식' 컬럼 삭제
6. 위에서부터 3개의 행만 선택 (loc 사용)
7. 위에서부터 3개의 행만 선택 (iloc 사용)
8. 주어진 데이터(df)에서 아래 값을 loc를 활용해 데이터프레임으로 출력
- 소나타 3500
- 클레오스 4000
9. 주어진 데이터(df)에서 아래 값을 iloc를 활용해 데이터프레임으로 출력
- 소나타 3500
- 클레오스 4000
10. 제품 중 가격이 가장 비싼 순으로 정렬해 상위 3개의 값 구하기

```python
# 1번: '연식' 컬럼의 결측치를 연식 데이터 중 최솟값으로 채우기
min_year = df['연식'].min()
df['연식'] = df['연식'].fillna(min_year)
df
```

```python
# 2번: '브랜드' 컬럼의 결측치를 브랜드 컬럼의 최빈값으로 채우기
mode_brand = df['브랜드'].mode()[0] # mode() 함수는 시리즈 형태로 반환하므로 최빈값만 얻고 싶으면 뒤에 인덱스 0 붙임.
df['브랜드'] = df['브랜드'].fillna(mode_brand)
df
```

```python
# 3번: 가격이 3500 이상인 데이터의 수 구하기
over_3500 = df['가격(만원)'] >= 3500
# len(df[over_3500])
sum(over_3500)
```

```python
# 4번: '이벤트가' 칼럼을 만들고 기존 가격에서 20% 할인된 가격 구하기
df['이벤트가'] = df['가격(만원)'] * 0.8
df
```

```python
# 5번: '연식' 컬럼 삭제
df = df.drop('연식', axis=1)
df
```

```python
# 6번: 위에서부터 3개의 행만 선택 (loc 사용)
df.loc[:2] # 인덱스명 또는 칼럼명을 통해 접근 (끝 포함)
```

```python
# 7번: 위에서부터 3개의 행만 선택 (iloc 사용)
df.iloc[:3] # iloc 슬라이싱 범위는 마지막 인덱스 번호 앞까지 결과 반환 (끝 미포함)
```

```python
# 8번: 주어진 데이터(df)에서 아래 값을 loc를 활용해 데이터프레임으로 출력
df.loc[1:2, :'가격(만원)']
```

```python
# 9번: 주어진 데이터(df)에서 아래 값을 iloc를 활용해 데이터프레임으로 출력
df.iloc[1:3, :2]
```

```python
# 10번: 제품 중 가격이 가장 비싼 순으로 정렬해 상위 3개의 값 구하기
df.sort_values(by='가격(만원)', ascending=False).head(3)
```
